```python
Df = pd.read_excel("../data/data_WHO.xlsx",header=1)
Df = Df[pd.notna(Df.date)] # revome the last row with total numbers
Df.date = pd.to_datetime(Df.date, format='%Y-%m-%d')
Df = pd.wide_to_long(Df,["confirmed","probable"],i="date",j="epicurve",sep='_').reset_index()
Df.epicurve = pd.to_datetime(["20"+str(x) for x in Df.epicurve], format='%Y%m%d')
Df = Df.loc[Df.epicurve>=Df.date]#.fillna(0)
Df[:6]
```

# Loading the dataset of daily counts from MOH

```python
df_MOH_reported = pd.read_excel("../data/data_WHO.xlsx",sheet_name="Rerorted_incidence_daily")
df_MOH_reported.report_date = pd.to_datetime(df_MOH_reported.report_date, format='%Y-%m-%d')
df_MOH_reported.confirmed_reported_incidence = df_MOH_reported.confirmed_reported_incidence.astype('int')
df_MOH_reported.probable_reported_incidence = df_MOH_reported.probable_reported_incidence.astype('int')
df_MOH_reported[:5]
```

```python
df_MOH_reported = df_MOH_reported.loc[:,['report_date','confirmed_reported_incidence','probable_reported_incidence']]\
    .rename(columns={'report_date':'date','confirmed_reported_incidence':'confirmed','probable_reported_incidence':'probable'})\
    .sort_values('date')
df_MOH_reported['weekday'] = df_MOH_reported.date.dt.weekday
df_MOH_reported[:5]
```

```python
# selecting only Sundays and calculating weekly incidence
df_MOH_reported.confirmed = df_MOH_reported.confirmed.cumsum()
df_MOH_reported.probable = df_MOH_reported.probable.cumsum()
df_MOH_reported = df_MOH_reported.loc[lambda d: d.weekday==6]
df_MOH_reported.confirmed = (df_MOH_reported.confirmed.diff(periods=1).fillna(df_MOH_reported.confirmed)).astype('int')
df_MOH_reported.probable = (df_MOH_reported.probable.diff(periods=1).fillna(df_MOH_reported.probable)).astype('int')
df_MOH_reported = df_MOH_reported.drop('weekday',1).reset_index(drop=True)
df_MOH_reported[:5]
```

## <font color="orange">Getting two "convenient" processed dataframes</font>

```python
def make_df_and_df0(Df, epicurve0):
    df0 = Df.loc[Df['epicurve']==epicurve0].copy()
    df = df0.loc[:,['epicurve','date','confirmed','probable','T']].\
        rename(columns={'confirmed':'confirmed0','probable':'probable0','epicurve':'epicurve0','T':'T0'}).\
        merge(Df.loc[Df.epicurve<epicurve0,['epicurve','date','confirmed','probable','t','T']],on='date',how='right').\
        loc[:,['date','epicurve','T','t','confirmed','probable','epicurve0','T0','confirmed0','probable0']]
    df.sort_values(['epicurve','date'],inplace=True)

    df['δt'] = df['T']-df['t']
    df['δt0'] = df['T0']-df['t']
    df['logδt'] = np.log(df['δt'])
    df['logδt0'] = np.log(df['δt0'])
    df['total'] = df0.confirmed+df0.probable

    df0['δt'] = df0['T']-df0['t']
    df0['logδt'] = np.log(df0['δt'])
    df0['total'] = df0.confirmed+df0.probable
    
    df = df.reset_index(drop=True)
    df0 = df0.reset_index(drop=True)
    
    return df, df0
```
